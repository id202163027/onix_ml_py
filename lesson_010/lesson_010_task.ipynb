```python
import scipy.integrate
import numpy as np
from numpy import sin,cos,tan
import matplotlib.pyplot as plt
import inspect
```

```python
# Trapezoid class with integral method and method, that plots the function f(x) in a given range [a,b]
class Trapezoid:
    def __init__(self, f, a, b):
        self.f = f # lambda func
        self.a = a # left lim
        self.b = b # right lim
    def integral(self):
        return scipy.integrate.quad(self.f, self.a, self.b)[0]
    def display(self):
        x = np.linspace(self.a, self.b, 100)
        dx = (self.b - self.a)/10
        x_ = np.linspace(self.a - dx, self.b + dx, 100)
        # lambda func code for label:
        f_code = inspect.getsource(self.f).strip().split(':')[-1]
        plt.plot(x_,self.f(x_),label=f_code)
        plt.fill_between(x,self.f(x),alpha=0.3)
        plt.vlines((self.a,self.b), 0, (self.f(self.a),self.f(self.b)), linestyles='dotted', colors='r')
        plt.legend()
        plt.show
```

```python
f = lambda x: sin(x) + cos(2*x)
a,b = 0, 1.2
```

```python
obj1 = Trapezoid(f,a,b)
print('Definite integral: %.3f' % obj1.integral())
obj1.display()
```

```python
f = lambda x: x*tan(x) + 2
a,b = 0.2, 0.8
```

```python
obj2 = Trapezoid(f,a,b)
print('Definite integral: %.3f' % obj2.integral())
obj2.display()
```

```python
f = lambda x: x**3 - 10*x**2 + 170
a,b = 0, 10
```

```python
obj3 = Trapezoid(f,a,b)
print('Definite integral: %.3f' % obj3.integral())
obj3.display()
```
