```python
import pandas as pd
```

#### 1. Load data from dataset.csv into a DataFrame.

```python
columns = ['Rank', 'Name', 'Platform', 'Year', 'Genre', 'Publisher',
           'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
```

```python
df = pd.read_csv('dataset.csv', names=columns, skiprows=1)
```

```python
df.tail()
```

#### 2. Display a given number of consecutive rows from any given position in the dataframe.

```python
def slice(df,start,n):
    return df.iloc[range(start, min(len(df), start + n))]
```

```python
slice(df,16595,4)
```

#### 3. Finds all unique genres (Genre) in the source data

```python
genres = pd.unique(df['Genre'])
print(genres)
```

#### 4-6. Separate dataframe into genre. Order rows in each genre in ascending order of the values in the Year column. Make re-index. Saves results to csv files.

```python
frames = {}
for genre in genres:
    frames[genre] = df[df['Genre'] == genre].sort_values('Year').reset_index(drop=True)
    frames[genre].to_csv(genre.lower() + ".csv")
```

#### Alternative way with groupby:

```python
frames2 = {name:group for name, group in df.groupby('Genre')}
```

```python
genres2 = list(frames2.keys())
print(genres2)
```

```python
for name,group in frames2.items():
    frames2[name] = group.sort_values('Year').reset_index(drop=True)
    frames2[name].to_csv(name.lower() + ".csv")
```
