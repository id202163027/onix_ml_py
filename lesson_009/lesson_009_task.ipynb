```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'x_1': ['A', 'B', 'C'] * 3, 'y_1': ['D', 'E', 'F'] * 3,
                    'z': np.random.randint(0, 20, 9)})
```

```python
df2 = pd.DataFrame({'factor1': ['A', 'B', 'C'] * 4,'factor2': ['D', 'E'] * 6,
                    'result': np.random.randint(0, 20, 12)})
```

```python
df1
```

```python
# rename df2 columns
df2_columns_new = ['x_2','y_2','z']
df2.rename(columns=dict(zip(df2.columns,df2_columns_new)), inplace=True)
df2
```

```python
# inner join
df3 = pd.merge(df1, df2, left_on=['x_1','y_1'], right_on=['x_2','y_2'], suffixes=('_left', '_right'), how='inner')
df3
```

```python
# outer join
df4 = df1.merge(df2, left_on=['x_1','y_1'], right_on=['x_2','y_2'], suffixes=('_left', '_right'), how='outer')
df4
```

```python
#df4.fillna(0, inplace=True)
```

```python
# list of NaN positions
nan_pos = df4.isnull().stack()[lambda x: x].index.tolist()
```

```python
# first 5 sample
nan_pos[:5]
```

```python
# set NaN to zero
for x,y in nan_pos:
    df4.loc[x,y] = 0
```

```python
df4
```

```python
# add on the right average of z_left, z_right
df4 = pd.concat([df4, df4[['z_left','z_right']].mean(axis=1).rename('z_avg')], axis=1)
# df4['z_avg'] = df4[['z_left','z_right']].mean(axis=1) # alternative
```
