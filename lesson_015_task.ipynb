```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# load 1-st image data
img_01_raw = tf.io.read_file('images/01.jpg')
```

```python
# decode 1-st image to a uint8 tensor
img_01 = tf.io.decode_jpeg(img_01_raw)
```

```python
# load 2-nd image data
img_02_raw = tf.io.read_file('images/02.jpg')
```

```python
# decode 2-nd image to a uint8 tensor
img_02 = tf.io.decode_jpeg(img_02_raw)
```

```python
# combine images
add_result = tf.add(img_01, img_02)
```

```python
# save image stored as a numpy array to a path
tf.keras.utils.save_img('images/add_result.jpg', add_result.numpy())
```

```python
# display source images and result
plt.figure(figsize=(16,16))
for i,image in enumerate([img_01,img_02,add_result]):
    plt.subplot(131 + i)
    plt.imshow(image)
plt.show()
```
