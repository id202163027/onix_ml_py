```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'x_1': ['A', 'B', 'C'] * 3, 'y_1': ['D', 'E', 'F'] * 3,
                    'z': np.random.randint(0, 20, 9)})
```

```python
df2 = pd.DataFrame({'factor1': ['X', 'Y', 'Z'] * 4,'factor2': ['F', 'G'] * 6,
                    'result': np.random.randint(0, 10, 12)})
```

```python
# rename df2 columns
df2_columns_new = ['x_2','y_2','z']
df2.rename(columns=dict(zip(df2.columns,df2_columns_new)), inplace=True)
```

```python
print(df1)
```

```python
print(df2)
```

```python
# inner join on keys ['x_1','y_1'] and ['x_2','y_2']
df3 = pd.merge(df1, df2, left_on=['x_1','y_1'], right_on=['x_2','y_2'], suffixes=('_left', '_right'))
print(df3)
```

```python
# inner join on keys y_1, y_2
df3 = df1.merge(df2, left_on=['y_1'], right_on=['y_2'], suffixes=('_left', '_right'), how='inner')
print(df3)
```

```python
# outer join on keys ['x_1','y_1'] and ['x_2','y_2']
df4 = df1.merge(df2, left_on=['x_1','y_1'], right_on=['x_2','y_2'], suffixes=('_left', '_right'), how='outer')
print(df4)
```

```python
# outer join on keys y_1, y_2
df4 = df1.merge(df2, left_on=['y_1'], right_on=['y_2'], suffixes=('_left', '_right'), how='outer')
print(df4)
```

```python
# replace all NaN elements with zero
df4.fillna(0, inplace=True)
print(df4)
```

```python
# add on the right average of z_left, z_right
df4 = pd.concat([df4, df4[['z_left','z_right']].mean(axis=1).rename('z_avg')], axis=1)
# df4['z_avg'] = df4[['z_left','z_right']].mean(axis=1) # alternative
print(df4)
```
