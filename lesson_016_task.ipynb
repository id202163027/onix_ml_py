```python
import numpy as np
from sklearn.neural_network import MLPClassifier
```

```python
# load initial dataset
base_dataset = np.loadtxt('base_dataset.csv', delimiter=',')
```

```python
print("A, B, C, D")
print(base_dataset)
```

```python
# construct additional dataset by repeating base_dataset 9 times
add_set = np.tile(base_dataset, (9, 1))
```

```python
# replace A,B,C zeros with random values in interval 0,0.4
np.place(add_set[:,:-1], add_set[:,:-1]==0, np.random.uniform(0,0.4,(add_set[:,:-1]==0).sum()))
```

```python
# replace A,B,C ones with random values in interval 0.6,1.2
np.place(add_set[:,:-1], add_set[:,:-1]==1, np.random.uniform(0.6,1.2,(add_set[:,:-1]==1).sum()))
```

```python
# create train dataset concatenating initial base_dataset and generated add_set
dataset = np.concatenate((base_dataset,add_set), axis=0)
```

```python
print("size:\n", dataset.shape)
print("dataset:\n", dataset[:16,:])
```

```python
# input and output for classifier
X = dataset[:,:-1]
y = dataset[:,-1]
```

```python
# initializing the MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
```

```python
# fitting data to the network
clf.fit(X, y)
```

```python
# generate data for predict
samples = np.concatenate((np.random.uniform(0.,0.4, 15),np.random.uniform(0.6,1.2, 15)))
np.random.shuffle(samples)
samples = samples.reshape((10,3))
```

```python
print(samples)
```

```python
clf.predict(samples)
```
