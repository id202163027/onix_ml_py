```python
import csv
import matplotlib.pyplot as plt
```

```python
# load data from file with csv.reader without using numpy
x = []; y = []
with open('dataset.txt', 'r') as file:
    reader = csv.reader(file, delimiter = ' ')
    next(reader)
    for row in reader:
        x.append(float(row[0]))
        y.append(float(row[1]))
```

```python
# find max point (x, y) for annotation
max_y = max(y)
max_x = x[y.index(max_y)]
```

```python
# plot linear chart y = f(x) with max annotation
plt.figure(figsize=(16,5))
plt.plot(x,y, color='#1f77b4', marker='.', linestyle='solid')
plt.title('Linear chart y = f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.annotate('max y = %.1f' %max_y, xy= (max_x,max_y),  xycoords='data',
            xytext=(0.6, 0.95), textcoords='axes fraction',
            bbox=dict(boxstyle="round", fc="0.9"),
            arrowprops=dict(arrowstyle="->"),
            horizontalalignment='right', verticalalignment='top',)
plt.axhline(y=0, color='r', alpha=0.5)
plt.grid(alpha=0.3)
plt.show()
```

```python
# plot stem y = f(x)
plt.figure(figsize=(16,5))
plt.title('Stem plot y = f(x)')
plt.stem(x, y)
plt.grid(axis = 'y', alpha=0.3)
plt.show()
```

```python
# plot step y = f(x)
plt.figure(figsize=(16,5))
plt.title('Step plot y = f(x)')
plt.step(x, y, where='mid')
plt.grid(axis = 'y', alpha=0.3)
plt.axhline(y=0, color='r', alpha=0.5)
plt.show()
```

```python
# plot bar y = f(x)
plt.figure(figsize=(16,5))
plt.title('Bar plot y = f(x)')
colors = ['#1f77b4' if i >= 0 else '#ff000090' for i in y] # default blue #1f77b4 and red hex RGBA
plt.bar(x, y, color=colors)
plt.grid(axis = 'y', alpha=0.3)
plt.show()
```

```python
# plot y1 = f(x) and y2 = 2*f(x)
y2 = [i*2 for i in y]
plt.figure(figsize=(16,5))
plt.plot(x, y2, color='grey', marker='o', linestyle='dashed', label='y2 = 2*f(x)')
plt.plot(x, y, color='#1f77b4', marker='.', linestyle='solid', label='y1 = f(x)')
plt.axhline(y=0, color='r', alpha=0.5)
plt.title('Linear chart y1, y2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
